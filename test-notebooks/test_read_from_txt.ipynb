```python


book_file = Path(r'/Volumes/2TB/Development/family-altar-book-processor/Book1_days_formatted.txt')
output_folder = Path("processed_books")
output_folder.mkdir(exist_ok=True)  # Create folder if it doesn't exist
output_file = output_folder / "text-file-creation-test.txt"

def read_book(file_path: Path) -> str:
    """Read the book text into a string, auto-detecting encoding"""
    raw_data = file_path.read_bytes()
    result = chardet.detect(raw_data)
    encoding = result['encoding']
    confidence = result['confidence']
    print(f"Detected encoding: {encoding} (confidence: {confidence:.2f})")
    text = raw_data.decode(encoding).strip()
    # Remove extra blank lines
    lines = [line.strip() for line in text.splitlines() if line.strip()]
    return "\n".join(lines)

if __name__ == "__main__":
    if book_file.exists():
        text = read_book(book_file)
        with output_file.open("w", encoding="utf-8") as f:
            f.write(text)
    else:
        print(f"File not found: {book_file}")
```

```python
def load_appendix(appendix_path: Path) -> dict[int, tuple[str, str]]:
    """
    Reads appendix.txt and returns:
    { day_number: (title, date) }
    """
    appendix = {}

    with open(appendix_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            # Format:
            # 1\tCountdown\t\tSeptember 9, 1962
            parts = re.split(r"\t+", line)

            if len(parts) >= 3:
                day_number = int(parts[0])
                title = parts[1].strip()
                date = parts[2].strip()
                appendix[day_number] = (title, date)

    return appendix
```

```python
import re
from pathlib import Path
from typing import Dict, Tuple

MONTHS = (
    "January|February|March|April|May|June|July|August|September|October|November|December"
)

# 1) Heading boundary matcher (finds start of each day section)
DAY_HEADING_RE = re.compile(
    rf"""
    (?P<full_heading>
        Day\s+(?P<day_number>\d+)\s*     # Day 1
        The\s+Family\s+Altar\s*          # The Family Altar
        (?P<month>{MONTHS})\s+           # January
        (?P<day_in_month>\d{{1,2}})      # 1
    )
    """,
    re.IGNORECASE | re.VERBOSE,
)

# 2) Appendix line parser: "1\tTitle\t\tSeptember 9, 1962"
APPENDIX_LINE_RE = re.compile(
    r"^\s*(\d+)\s*\t+([^\t]+?)\s*\t+(.+?)\s*$"
)


def load_appendix(appendix_path: Path) -> Dict[int, Tuple[str, str]]:
    """
    Returns: { day_number: (title, date) }
    """
    appendix: Dict[int, Tuple[str, str]] = {}

    with open(appendix_path, "r", encoding="utf-8") as f:
        for raw_line in f:
            line = raw_line.rstrip("\n")
            if not line.strip():
                continue

            m = APPENDIX_LINE_RE.match(line)
            if not m:
                # Skip anything that doesn't match expected format
                continue

            day_number = int(m.group(1))
            title = m.group(2).strip()
            date = m.group(3).strip()

            appendix[day_number] = (title, date)

    return appendix


def save_days_to_files(book_text: str, output_folder: Path, appendix_data: Dict[int, Tuple[str, str]]):
    output_folder.mkdir(exist_ok=True)

    matches = list(DAY_HEADING_RE.finditer(book_text))

    for i, match in enumerate(matches):
        start = match.start()
        end = matches[i + 1].start() if i + 1 < len(matches) else len(book_text)

        day_text = book_text[start:end].strip()

        day_number = int(match.group("day_number"))
        month = match.group("month").capitalize()
        day_in_month = int(match.group("day_in_month"))

        filename = f"{month}_{day_in_month}.txt"

        # Append title + date at the end of the daily reading
        appendix_entry = appendix_data.get(day_number)
        if appendix_entry:
            title, date = appendix_entry
            day_text += f"\n\n[[[{title} — {date}]]]"
        else:
            # Optional: keep silent, or leave a marker for debugging
            # day_text += f"\n\n[Missing appendix entry for Day {day_number}]"
            pass

        output_path = output_folder / filename
        output_path.write_text(day_text, encoding="utf-8")

        print(f"Saved: {filename} (Day {day_number})")


# ---- Replace these with your existing functions/paths ----
def read_book(book_file: Path) -> str:
    return book_file.read_text(encoding="utf-8")


if __name__ == "__main__":
    book_file = Path(r'/Volumes/2TB/Development/family-altar-book-processor/Book1_days_formatted.txt')
    output_folder = Path("daily_readings")
    appendix_file = Path('/Volumes/2TB/Development/family-altar-book-processor/appendix.txt')

    if not book_file.exists():
        print(f"File not found: {book_file}")
        raise SystemExit(1)

    if not appendix_file.exists():
        print(f"File not found: {appendix_file}")
        raise SystemExit(1)

    text = read_book(book_file)
    print(len(text))
    appendix_data = load_appendix(appendix_file)
    print(len(appendix_data))

    save_days_to_files(text, output_folder, appendix_data)
```
